```python
file_path = 'data/Capstone Info1.xlsx'
xl = pd.ExcelFile(file_path)

od_deaths_df = pd.read_excel(file_path, sheet_name='all drug OD Deaths')

# Clean the data: Replace any NaN values with 0 (assuming NaN means no deaths recorded)
od_deaths_cleaned_df = od_deaths_df.fillna(0)
od_deaths_long_df = pd.melt(od_deaths_cleaned_df, id_vars=['Unnamed: 0'], var_name='Year', value_name='Deaths')

od_deaths_long_df.rename(columns={'Unnamed: 0': 'Category'}, inplace=True)

od_deaths_long_df.head()
```

```python
# Save the cleaned and reshaped data to a new CSV file
output_file_path = 'data/Cleaned_OD_Deaths_Data.csv'
od_deaths_long_df.to_csv(output_file_path, index=False)

output_file_path
```
