```python
import pandas as pd
import geopandas as gpd
import zipfile
import matplotlib.pyplot as plt

# Load the overdose data
data_path = 'data/data-table.csv'
data = pd.read_csv(data_path)

# Unzip and load the geojson file for US states
zip_path = 'data/geo-admin1-us_zip.zip'
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall('/mnt/data/')
geojson_path = 'data/admin1-us.geojson'
geo_df = gpd.read_file(geojson_path)

# Merge the overdose data with the geo data frame based on state codes
merged_df = geo_df.merge(data, left_on='state_code', right_on='STATE')

# Filter the data for the year 2021
data_2021 = merged_df[merged_df['YEAR'] == 2021]

# Plotting the density heat map for overdose death in the United States for 2021
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
data_2021.plot(column='RATE', cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)
ax.set_title('Density Heat Map of Overdose Death Rates in the United States, 2021')
plt.show()
```

```python
import numpy as np

# Creating a spatial heat map requires a different approach. We'll use the RATE column for the heat map.
# First, we need to ensure that the RATE column is in a numeric format.
data_2021['RATE'] = pd.to_numeric(data_2021['RATE'], errors='coerce')

# Replace NaN values with 0 for visualization purposes
data_2021['RATE'].fillna(0, inplace=True)

# We will use the centroid of each state for plotting the heat map.
data_2021['centroid'] = data_2021.geometry.centroid
centroids = data_2021.set_geometry('centroid')

# Plotting the spatial heat map
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
base = geo_df.boundary.plot(ax=ax, color='black')
centroids.plot(ax=base, marker='o', column='RATE', cmap='OrRd', 
               markersize=np.sqrt(centroids['RATE'])*10, legend=True, alpha=0.6)
ax.set_title('Spatial Heat Map of Overdose Death Rates in the United States, 2021')
plt.show()
```
