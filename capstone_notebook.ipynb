```python

file_path_2019_2022 = 'data/nationwide-drugs-fy19-fy22.csv'
file_path_2020_2023 = 'data/nationwide-drugs-fy20-fy23.csv'

data_2019_2022 = pd.read_csv(file_path_2019_2022)
data_2020_2023 = pd.read_csv(file_path_2020_2023)

data_2019_2022.head(), data_2020_2023.head()
```

```python

merged_data = pd.concat([data_2019_2022, data_2020_2023])

total_quantity_by_drug = merged_data.groupby('Drug Type')['Sum Qty (lbs)'].sum().sort_values(ascending=False)

events_by_drug = merged_data.groupby('Drug Type')['Count of Event'].sum().sort_values(ascending=False)

monthly_trends = merged_data.groupby(['FY', 'Month (abbv)'])['Sum Qty (lbs)'].sum().reset_index()

regional_variation = merged_data.groupby(['Region', 'Drug Type'])['Sum Qty (lbs)'].sum().reset_index()

plt.figure(figsize=(15, 6))

# Plot for total quantity by drug
plt.subplot(2, 2, 1)
sns.barplot(x=total_quantity_by_drug.index, y=total_quantity_by_drug.values)
plt.xticks(rotation=45)
plt.title('Total Quantity of Each Drug Type Seized (2019-2023)')
plt.ylabel('Total Quantity (lbs)')
plt.xlabel('Drug Type')

# Plot for number of events per drug type
plt.subplot(2, 2, 2)
sns.barplot(x=events_by_drug.index, y=events_by_drug.values)
plt.xticks(rotation=45)
plt.title('Number of Events Per Drug Type')
plt.ylabel('Count of Events')
plt.xlabel('Drug Type')

# Plot for monthly trends
plt.subplot(2, 2, 3)
sns.lineplot(data=monthly_trends, x='Month (abbv)', y='Sum Qty (lbs)', hue='FY')
plt.title('Monthly Drug Seizure Trends (2019-2023)')
plt.ylabel('Total Quantity (lbs)')
plt.xlabel('Month')

# Plot for regional variations
plt.subplot(2, 2, 4)
sns.barplot(x='Sum Qty (lbs)', y='Drug Type', hue='Region', data=regional_variation)
plt.title('Regional Variations in Drug Seizures')
plt.ylabel('Drug Type')
plt.xlabel('Total Quantity (lbs)')

plt.tight_layout()
plt.show()
```

```python

opioids_list = ['Fentanyl', 'Heroin', 'Other Opioids']
opioid_data = merged_data[merged_data['Drug Type'].isin(opioids_list)]

# Total quantity of opioids seized over the years
opioid_quantity_by_year = opioid_data.groupby(['Drug Type', 'FY'])['Sum Qty (lbs)'].sum().reset_index()

# Trends in opioid seizures over the months
opioid_monthly_trends = opioid_data.groupby(['Drug Type', 'FY', 'Month (abbv)'])['Sum Qty (lbs)'].sum().reset_index()

# Regional variations in opioid seizures
opioid_regional_variation = opioid_data.groupby(['Region', 'Drug Type'])['Sum Qty (lbs)'].sum().reset_index()

plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.barplot(x='FY', y='Sum Qty (lbs)', hue='Drug Type', data=opioid_quantity_by_year)
plt.title('Total Quantity of Opioids Seized by Year')
plt.ylabel('Total Quantity (lbs)')
plt.xlabel('Fiscal Year')

plt.subplot(1, 3, 2)
sns.lineplot(data=opioid_monthly_trends, x='Month (abbv)', y='Sum Qty (lbs)', hue='Drug Type')
plt.title('Monthly Opioid Seizure Trends')
plt.ylabel('Total Quantity (lbs)')
plt.xlabel('Month')

plt.subplot(1, 3, 3)
sns.barplot(x='Sum Qty (lbs)', y='Drug Type', hue='Region', data=opioid_regional_variation)
plt.title('Regional Variations in Opioid Seizures')
plt.ylabel('Drug Type')
plt.xlabel('Total Quantity (lbs)')

plt.tight_layout()
plt.show()
```
